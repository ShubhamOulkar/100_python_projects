```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("https://web.archive.org/web/20200518073855/https://www.empireonline.com/movies/features/best-movies-2/")
movie = response.text
```

```python
soup = BeautifulSoup(movie, 'html.parser')
print(soup)
```

```python
movie_name = soup.find_all(name="h3", class_="title")
```

```python
movie_names = [movie_name[i].text for i in range(len(movie_name))]
print(movie_names)
```

```python
for i in range(len(movie_names)):
  with open("100_movies.txt", "a") as file:
    file.write("\n" + str(movie_names[len(movie_names)-1-i]))

print("100_movies.txt file is ready")
```
